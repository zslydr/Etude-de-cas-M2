```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
img=mpimg.imread('C:/Users/Pierre Lavigne/Dropbox/Etude de Cas M2/tilling.jpg')
imgplot = plt.imshow(img)
plt.show()
```

### Start the algorithm of region tilling

```python
imgtest=mpimg.imread('C:/Users/Pierre Lavigne/Dropbox/Etude de Cas M2/tillingtest.jpg')
plt.imshow(imgtest)
plt.show()
```

#### Region to till:


```python
x,y,z = np.shape(img)
Im = imgtest-img 
imgplot = plt.imshow(Im)
plt.show()

measure = np.zeros((x,y))
summeasure = 0
for i in range(x):
    for j in range(y):
        measure[i][j] = sum(Im[i][j][k] for k in range(2))
print(measure)
```

```python
Omega = measure.nonzero()
Omega
plt.imshow(measure)
plt.show()
```

### Let's play a bit with measures, we apply first a random noise on the initial image

## A first naive way to solve the problem, fill the gap only by taking the average around the region to till

```python
for (i,j) in zip(Omega[0],Omega[1]):
    if (i + j)%2:
        img[i][j] = img[i-1-round(np.random.uniform(0,10))][j] 
    else:
        img[i][j] = img[i][j-round(np.random.uniform(0,10))-1]
    
```

```python
plt.imshow(img)
plt.show()
```
