```python
def cosmat(imag1,imag2):
    mn = range(imag1.shape[0])
    matrix = [[0 for i in mn] for j in mn]
    for (i,j) in zip(mn,mn):
        print(i,j)
        matrix[i][j] = np.dot(imag1[i],imag2[j])/(norm(imag1[i])*norm(imag2[j])+1)
    return matrix
```
